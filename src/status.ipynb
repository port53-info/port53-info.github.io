```python
#本代码结果用于替换/content/data/Open encrypted DNS Servers/index.md中的内容；替换对应变量名的内容
## 后续可以增加一个匹配和自动替换功能
import json
from pprint import pprint

with open("result_ip_count.json", "r") as inf:
    ip_count = json.loads(inf.read())
with open("result_adn_count.json","r") as inf:
    adn_count = json.loads(inf.read())

dot = []
for k,v in ip_count['dot'].items():
    dot.append([k,v,adn_count['dot'][k]])
dot = sorted(dot, key=lambda x: x[0])
print("const dot = %s;"%str(dot))

doh = []
for k,v in ip_count['doh'].items():
    doh.append([k,v,adn_count['doh'][k]])
doh = sorted(doh, key=lambda x: x[0])
print("const doh = %s;"%str(doh))

doq = []
for k,v in ip_count['doq'].items():
    doq.append([k,v,adn_count['doq'][k]])
doq = sorted(doq, key=lambda x: x[0])
print("const doq = %s;"%str(doq))

doh3 = []
for k,v in ip_count['doh3'].items():
    doh3.append([k,v,adn_count['doh3'][k]])
doh3 = sorted(doh3, key=lambda x: x[0])
print("const doh3 = %s;"%str(doh3))
```

```python
import json
from pprint import pprint

with open("result_cert_validity.json", "r") as inf:
    d = json.loads(inf.read())
# pprint(d)
valid, self_signed = [],[]

for k,v in d['dot'].items():
    t = []
    t.append(k)
    t.append(d['dot'][k]['valid'])
    if k in d['doh']:
        t.append(d['doh'][k]['valid'])
    else:
        t.append(0)
    if k in d['doq']:
        t.append(d['doq'][k]['valid'])
    else:
        t.append(0)
    if k in d['doh3']:
        t.append(d['doh3'][k]['valid'])
    else:
        t.append(0)
    valid.append(t)
valid = sorted(valid, key=lambda x: x[0])

for k,v in d['dot'].items():
    t = []
    t.append(k)
    t.append(d['dot'][k]['self-signed'])
    if k in d['doh']:
        t.append(d['doh'][k]['self-signed'])
    else:
        t.append(0)
    if k in d['doq']:
        t.append(d['doq'][k]['self-signed'])
    else:
        t.append(0)
    if k in d['doh3']:
        t.append(d['doh3'][k]['self-signed'])
    else:
        t.append(0)
    self_signed.append(t)
self_signed = sorted(self_signed, key=lambda x: x[0])

print("const valid = %s;"%str(valid))
print("const self_signed = %s;"%str(self_signed))
```
